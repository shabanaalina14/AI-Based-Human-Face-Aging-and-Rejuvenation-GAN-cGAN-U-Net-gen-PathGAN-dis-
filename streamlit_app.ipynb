```python
import streamlit as st
from PIL import Image
import torch
from models.generator import UNetGenerator

model = UNetGenerator()
model.load_state_dict(torch.load("generator_fast.pth"))
model.eval()

st.title("ðŸ§“ AI Face Aging Generator")
img = st.file_uploader("Upload a face image")

age = st.slider("Select Target Age Group", 0, 100, 50)

if img:
    image = Image.open(img)
    st.image(image, caption="Original Face")
    fake_face = model(image, torch.tensor(age))
    st.image(fake_face, caption=f"Aged Face ({age})")
```
